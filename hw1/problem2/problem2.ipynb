# Problem 2
Use Standard Least Square to fit curve to the ball trajectory in videos.

## video1 (no noise)
1. Filter out the ball pixel position

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

cap = cv2.VideoCapture('ball_video1.mp4')

if (cap.isOpened() == False):
    print("Error opening video stream or file")

dots = []
heights = 0
widths = 0
while(cap.isOpened()):
    ret, frame = cap.read()

    if ret == True:
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY) # change frame image to gray scale to filter out the ball
        heights, widths = gray.shape
        ball_h = np.where(gray < 200)[0] # filter out pixels that are less white (background is white)
        ball_w = np.where(gray < 200)[1]
        if(ball_h.size!=0):
            h_center = (np.max(ball_h)+np.min(ball_h))/2 # calculate the average of the top, bottom value as the height center
            w_center = (np.max(ball_w)+np.min(ball_w))/2 # calculate the average of the left, right value as the width center
            xy_center = (w_center, heights - h_center -1) # transform h, w (0 is at the top left corner) to x, y (0 is at the bottom left corner)
            dots.append(xy_center)
    else:
        break
        
print("h, w: ", (heights, widths))
print("dots: ", len(dots))

fig, ax = plt.subplots(figsize=(7, 5))
plt.xlim([0, widths])
plt.ylim([0, heights])

for dot in dots:
    x, y = dot
    plt.scatter(x, y, 3, c='r')

cap.release()
plt.show()
```

2. Fit the ball trajectory with parabola curve ax^2+bx+c using standard least square.

```python
#assume the curve is a parabola that looks like ax^2+bx+c=0
x=[]
y=[]
for dot in dots:
    x.append(dot[0])
    y.append(dot[1])
Y = np.array([y]).T
X_col0 = np.square(np.array([x])).T 
X_col1 = np.array([x]).T
X_col2 = np.ones((len(x), 1))
X = np.hstack((X_col0, X_col1, X_col2))
B = np.matmul(np.linalg.inv(X.T.dot(X)), X.T.dot(Y)) # B = inv(X.T*X)(X.T*Y)
print(B)
```

```python
fig, ax = plt.subplots(figsize=(7, 5))
plt.xlim([0, widths])
plt.ylim([0, heights])
new_x = np.linspace(0, widths, 100)
new_y = B[0] * new_x**2 + B[1] * new_x + B[2]
plt.plot(new_x, new_y)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(21, 5))
ax1 = plt.subplot(1,3,1)
ax1.set_xlim([0, widths])
ax1.set_ylim([0, heights])
ax1.set_title("raw points")
for dot in dots:
    x, y = dot
    ax1.scatter(x, y, 3, c='r')
    
ax2 = plt.subplot(1,3,2)
ax2.set_xlim([0, widths])
ax2.set_ylim([0, heights])
ax2.set_title("curve")
ax2.plot(new_x, new_y)

ax3 = plt.subplot(1,3,3)
ax3.set_xlim([0, widths])
ax3.set_ylim([0, heights])
ax3.set_title("fit")
for dot in dots:
    x, y = dot
    ax3.scatter(x, y, 3, c='r')
ax3.plot(new_x, new_y)

plt.show()
```
